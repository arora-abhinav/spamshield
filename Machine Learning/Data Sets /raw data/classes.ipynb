```python
#Required libraries
import pandas as pd
from collections import defaultdict

#Loading datasets and obtaining columns based on names
combined_dataset = pd.read_csv("combined_dataset.csv")
spam = pd.read_csv("spam.csv", encoding='latin-1')
spam_classes = spam['v1'].values
combined_dataset_classes = combined_dataset['target'].values
final_dataset = pd.read_csv('final.csv')
final_dataset_classes = final_dataset['Class'].values


#Storing as default dicts
spam_class_count = defaultdict(int)
combined_class_count = defaultdict(int)
smish_class_count = defaultdict(int)
final_class_count = defaultdict(int)

#Obtaining actual class counts
for values in spam_classes:
    spam_class_count[values] += 1
for values in combined_dataset_classes:
    combined_class_count[values] += 1

for values in final_dataset_classes:
    final_class_count[values] += 1

with open('SMSSmishCollection.txt', 'r') as file:
    contents = file.readlines()
    for line in contents:
        c = line.split()[0]
        smish_class_count[c] += 1


#Combining all datasets
final = defaultdict(int)
final.update(smish_class_count)
final.update(spam_class_count)
final.update(combined_class_count)
final['spam'] += final['smish']
final.pop("smish")
print(final_class_count)

#Final class imbalance before synthetic data generation:
#{'ham': 8555, 'spam': 3153}
```
